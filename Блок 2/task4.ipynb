Задача 4.
Симметричные пары:
Условие: Дана таблица items (id, name, category).
Задание: Найди все пары записей, у которых полностью совпадают name и category. Пары (id1, id2) и (id2, id1) считай идентичными (оставь только одну). Пары с одинаковыми ID необходимо исключить.
Ожидаемый результат: id1, id2, category.

```python
import pandas as pd
```

```python
data = {
    'name': ['Товар А', 'Товар Б', 'Товар А', 'Товар А', 'Товар Б'],
    'category': ['Электроника', 'Одежда', 'Электроника', 'Книги', 'Одежда']
}
items = pd.DataFrame(data)
items = items.reset_index().rename(columns={'index': 'id'})
```

```python
items
```

```python
merged = pd.merge(items, items, on=['name', 'category'], suffixes=('1', '2'))
```

```python
merged
```

```python
result = merged[merged['id1'] < merged['id2']] # исключение идентичных пар и записей без пары
result = result[['id1', 'id2', 'category']]
result
```
