Задача 2.
Цепочки ID:
Условие: Даны две таблицы: users (id, new_id) и links (id1, id2), описывающие историю изменений ID одного пользователя (например, 1 $\rightarrow$ 2 $\rightarrow$ 3).
Используй два подхода к решению: рекурсивный и нет.
Задание: На основе связей определи для каждого исходного ID единый финальный new_id (минимальный ID в цепочке) и дедуплицируй записи.
Ожидаемый результат: id -> new_id (например, 1, 2, 3, 5 -> 1; 4 -> 4; 7, 8 -> 7).

```python
import pandas as pd
import networkx as nx

users = pd.DataFrame({
    'id': [1, 2, 3, 5, 6, 7, 10, 11, 15, 16, 20, 21, 25, 30, 31, 32, 40, 41, 50, 51, 60],
    'new_id': [2, 3, 4, 6, 7, 8, 11, 12, 16, 17, 21, 22, 26, 31, 32, 33, 41, 42, 51, 52, 61]
})

links = pd.DataFrame({
    'id1': [1, 2, 3, 5, 6, 7, 10, 11, 15, 16, 20, 21, 25, 30, 31, 32, 40, 41, 50, 51, 60],
    'id2': [2, 3, 4, 6, 7, 8, 11, 12, 16, 17, 21, 22, 26, 31, 32, 33, 41, 42, 51, 52, 61]
})

print(users.head(3))
print(links.head(3))
```

```python
G = nx.from_pandas_edgelist(links, source='id1', target='id2')
G.add_nodes_from(users['id'])
mapping = {}
for component in nx.connected_components(G):
    print(component)
    min_id = min(component)
    for node in component:
        mapping[node] = min_id
```

```python
result = pd.DataFrame(list(mapping.items()), columns=['id', 'new_id'])
result
```
