Задача 3.
Очистка строк:
Условие: Дана таблица data (id, name). В поле name перемешаны визуально одинаковые кириллические и латинские символы (а/а, о/о, с/с и т.д.).
Задание: Приведи имена к единому виду (к одной системе письма), затем найди дубликаты и присвой каждой группе дублей новый new_id.

```python
import pandas as pd
import random
```

# Создание датасета

```python
lat_to_cyrl = {
    'a': 'а',
    'o': 'о',
    'c': 'с',
    'e': 'е',
    'p': 'р',
    'x': 'х',
    'y': 'у',
    'i': 'і',
    'k': 'к',
    'T': 'Т',
    'B': 'В',
    'H': 'Н',
    'M': 'М',
    'A': 'А',
    'O': 'О',
    'C': 'С',
    'E': 'Е',
}

base_words = [
    'apple', 'banana', 'orange', 'grape', 'melon',
    'cherry', 'peach', 'plum','plum','plum','plum','plum','plum',
    'papaya', 'coconut', 'date', 'fig', 'lemon',
    'lime', 'nectarine', 'olive', 'pear', 'quince',
    'raspberry', 'strawberry', 'tangerine', 'watermelon',
    'cake', 'pie', 'cookie', 'bread', 'butter',
    'cheese', 'milk', 'yogurt', 'juice', 'tea',
    'coffee', 'water', 'soda', 'wine', 'beer',
    'cider', 'vodka', 'rum'
]

def mix_cyrillic(word: str) -> str:
    chars = list(word)
    replaceable = [(i, c) for i, c in enumerate(chars) if c in lat_to_cyrl]
    
    if not replaceable:
        return word
    
    num_to_replace = random.randint(1, min(3, len(replaceable)))
    chosen = random.sample(replaceable, num_to_replace)
    
    for idx, char in chosen:
        chars[idx] = lat_to_cyrl[char]
    
    return ''.join(chars)


data = []
for i in range(1, 43):
    word = base_words[i - 1]
    mixed = mix_cyrillic(word)
    data.append({'id': i, 'name': mixed})

df = pd.DataFrame(data)
```

# Решение задачи

```python
def to_lat(string):
    cyrl_to_lat = {value: key for key, value in lat_to_cyrl.items()}
    new_string = ''
    for n in string:
        if n in cyrl_to_lat.keys():
            new_string += cyrl_to_lat[n]
        else:
            new_string += n
    return new_string
```

```python
df['name'] = df['name'].apply(to_lat)
df.head(5)
```

```python
df['new_id'] = df.groupby('name').ngroup()
```
